```python
# Create a figure with 2 panels side by side
fig, axes = create_multi_panel_figure(nrows=1, ncols=2, figsize=(15, 6))

# Panel (a) - Simple charge balance
plot_charge_balance_simple_ax(
    axes[0], 
    analyzer, 
    formula='NaCo0.5Fe0.5O2', 
    alkali_metal='Na', 
    interval=0.01, 
    max_range=2.0
)
add_panel_label(axes[0], '(a)')

# Panel (b) - Substitution charge balance
plot_charge_balance_substitution_ax(
    axes[1],
    analyzer,
    base_formula="NaCoO2",
    metal_A="Co",
    metal_B="Fe",
    fractions=np.linspace(0.1, 0.9, 9),
    cmap_name="inferno"
)
add_panel_label(axes[1], '(b)')

plt.tight_layout()
plt.savefig('combined_charge_balance.png', dpi=600)  # Save the figure
plt.show()
```

```python
# Create a figure with 2 panels side by side
fig, axes = create_multi_panel_figure(nrows=1, ncols=2, figsize=(15, 6))

# Panel (a) - Average oxidation states
plot_avg_oxidation_states_ax(
    axes[0],
    analyzer, 
    formula='NaCo0.7Ni0.2Mn0.1O2', 
    alkali_metal='Na', 
    interval=0.015, 
    max_range=2.0, 
    colormap='inferno', 
    cmap_range=(0.2, 0.8)
)
add_panel_label(axes[0], '(a)')

# Panel (b) - Individual oxidation states
plot_individual_oxidation_states_ax(
    axes[1],
    analyzer, 
    formula='NaCo0.7Ni0.2Mn0.1O2', 
    alkali_metal='Na', 
    interval=0.015, 
    max_range=1.0, 
    colormap='inferno', 
    cmap_range=(0.1, 0.9)
)
add_panel_label(axes[1], '(b)')

plt.tight_layout()
plt.savefig('oxidation_states_combined.png', dpi=600)
plt.show()
```
