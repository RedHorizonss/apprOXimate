```python
from approximate import ApprOXimate

analyser = ApprOXimate()
```

```python
import pandas as pd

df = pd.read_csv('master_list_p2o3.csv')
df
```

```python
oximate_strings = []

for formula in df['Chemical']:
    result_string = analyser.charge_balance(formula, return_format='string')
    result_string = ";".join(result_string.split(";")[:-1])
    oximate_strings.append(result_string)

df['Oximate_Strings'] = oximate_strings
```

```python
df.to_csv('master_list_p2o3_with_oximate_strings.csv', index=False)
```
